```python
import pandas as pd
df = pd.read_excel(r'data/mice+protein+expression/Data_Cortex_Nuclear.xlsx')    # data recoded as xlsx

from sklearn.preprocessing import LabelEncoder

# Encode categorical columns and drop originals
columns_to_encode = ['MouseID', 'Genotype', 'Treatment', 'Behavior', 'class']

for col in columns_to_encode:
    if col in df.columns:
        le = LabelEncoder()
        df[f'{col}_encoded'] = le.fit_transform(df[col])
        print(f"{col} mapping:", dict(zip(le.classes_, le.transform(le.classes_))))
        # Drop the original column
        df = df.drop(columns=[col])

df
```

```python
from assumptions.ClassificationAssumptionsChecker import ClassificationAssumptionsChecker
checker = ClassificationAssumptionsChecker(df, target="class_encoded")

assumptions = checker.check_assumptions()
checker.report_assumptions()
checker.recommend_models()
```
