```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv(r'data/online+news+popularity/OnlineNewsPopularity.csv')

# Clean column names by removing leading/trailing whitespace
df.columns = df.columns.str.strip()

# Initialize LabelEncoder
le = LabelEncoder()

# Encode the 'url' column
df['url'] = le.fit_transform(df['url'])

print(f"URL column has been encoded with {len(le.classes_)} unique values")
print(f"First 10 encoded URL values: {df['url'].head(10).tolist()}")
print(f"Cleaned column names: {list(df.columns[:10])}...")  # Show first 10 column names

df
```

```python
from assumptions.RegressionAssumptionsChecker import RegressionAssumptionsChecker
checker = RegressionAssumptionsChecker(df, target="shares")  # Now using cleaned column name

checker.check_assumptions()
checker.report_assumptions()
```
