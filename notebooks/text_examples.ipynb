```python
def load_and_preprocess(path):
    """Load a JSONL file and split into train/test sets."""
    with open(path, "r", encoding="utf-8") as f:
        records = [json.loads(line) for line in f if line.strip()]
    
    df = pd.DataFrame(records)

    if 'Target' not in df.columns:
        raise ValueError("Missing 'Target' column for stratification.")

    return df

df = load_and_preprocess(r'C:\Users\user\ML\Classification\essays\train.jsonl')
df = df[['Text', 'Target']]

# Swap binary values: 0 becomes 1, 1 becomes 0
df['Target'] = df['Target'].replace({0: 1, 1: 0})

df
```
